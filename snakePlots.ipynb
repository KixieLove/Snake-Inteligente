```python
import pickle
import matplotlib.pyplot as plt

with open(r'D:\UAEM\4to semestre\Busqueda bayesiana\Codes\data1.pkl', 'rb') as f:
    data = pickle.load(f)
 
all_generations = data['all_generations']
best_individual = data['best_individual']
best_scores = data['best_scores']
avg_scores = data['avg_scores']
best_score = data['best_score']
best_fruits_eaten = data['best_fruits_eaten']
results = data['results']
scores = data['scores']
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10, 5))
plt.plot(best_scores, label='Best Scores')
plt.plot(avg_scores, label='Average Scores')
plt.xlabel('Generation')
plt.ylabel('Score')
plt.legend()
plt.title('Scores Over Generations')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(scores, bins=20)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Histogram of Scores From Last Generation')
plt.show()
```
